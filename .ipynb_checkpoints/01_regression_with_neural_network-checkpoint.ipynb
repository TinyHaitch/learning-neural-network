## Introduction to Regression with Neural Network

```python
import numpy as np 
import matplotlib.pyplot as plt 
import tensorflow as tf 
tf.__version__
```

```python
x= np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])
y= np.array([3., 6., 9., 12., 15., 18., 21., 24.])
```

```python
plt.plot(x, y)
```

## Input and Output Shape

```python
 y ==  x + 10
```

```python
## Turn our numpy arrays into tensor 
y = tf.constant(y, dtype=tf.float16)
x = tf.constant(x, dtype=tf.float16)

x, y
```

```python
## Steps in moderlling with tensorflow
# Creating a model, compiling the model, and fitting the model 

tf.random.set_seed(22)
## creating a model using secuential API 

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, activation='relu')
])

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(0.0001),
    metrics=['mae']
)

model.fit(tf.expand_dims(x, axis=1), y, epochs=100)
```

```python
## Try make prediction 
x_pred = tf.constant([17., 20., 23.], dtype=tf.float16)
model.predict(x_pred)
```

```python
x_data = tf.range(-200, 500, 3)
x_data
```

```python
y_data = x_data + 12
y_data
```

```python
len(x_data)
```

```python
x_data_train = x_data[:190]
y_data_train = y_data[:190]

x_data_test = x_data[190:]
y_data_test = y_data[190:]
```

```python
len(x_data_train), len(y_data_train), len(x_data_test), len(y_data_test)
```

```python
tf.random.set_seed(11)
m_model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=50, activation='relu'),
    tf.keras.layers.Dense(units=100, activation='relu'),
    tf.keras.layers.Dense(units=100, activation='relu'),
    tf.keras.layers.Dense(units=50, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='linear')
])

m_model.compile(
    loss = tf.keras.losses.mse,
    optimizer = tf.keras.optimizers.Adam(0.001),
    metrics = ['mse']
)

history = m_model.fit(tf.expand_dims(x_data_train, axis=1), y_data_train, epochs=150)
```

```python
y_pred =m_model.predict(x_data_test)
```

```python
import pandas as pd 
pd.DataFrame(history.history).plot()
plt.show()
```

```python
m_model.summary()
```

```python
def plot_predictions(train_data, train_labels, test_data, test_label, predictions):
    plt.figure(figsize=(12, 5.5))
    plt.scatter(train_data, train_labels, c='b', label='Training data')
    plt.scatter(test_data, test_label, c='g', label='Test data')
    plt.scatter(test_data, predictions, c='r', label='Predictions')
    plt.legend(); 
```

```python
y_pred, y_data_test
```

```python
plot_predictions(x_data_train, y_data_train,x_data_test, y_data_test, y_pred)
```

```python
m_model.evaluate(x_data_test, y_data_test)
```

```python
def mse(y_true, y_pred):
    mse = tf.metrics.mse(y_data_test, tf.squeeze(y_pred)) / 2
    return mse

def mae(y_true, y_pred): 
    mae = tf.metrics.mae(y_data_test, tf.squeeze(y_pred))
    return mae
```

```python
f'mean square error: {mse(y_data_test, y_pred)}', f'mean absolute error: {mae(y_data_test, y_pred)}'
```
