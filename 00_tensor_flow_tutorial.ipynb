```python
## Creating random tensors
random_1 = tf.random.set_seed(23);
random_1 = tf.random.normal([4,3])
random_2 = tf.random.set_seed(23);
random_2 = tf.random.normal([4,3])

random_1, random_2, random_1 == random_2
```

```python
## matrix multiplication 
tsf = tf.constant([[10, 7], 
                   [5, 7]])
```

```python
## Tesnor aggregation 

tf.random.set_seed(42)
F = tf.random.uniform(shape=[50], dtype=tf.float16, seed=42)
F
```

```python
tf.reduce_max(F)
```

```python
F[tf.argmax(F)]
```

```python
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
F[tf.argmin(F)]
```

```python
tf.argmin(F)
```
